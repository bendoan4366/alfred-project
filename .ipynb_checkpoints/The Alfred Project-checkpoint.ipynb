```python
spotify_UN = ("kanujbains")
spotify_PW = ("f8I9?weL")
song_stats = pd.read_csv('assets/data/songstats.csv', encoding='latin-1')
```

```python
song_stats.drop_duplicates(inplace = True)
song_stats.head(10)
```

```python
features_list = list(song_stats.columns)
features = song_stats.loc[:, features_list]
ax = features.hist(figsize=(20, 10), alpha = 0.6)
```

```python
song_stats.dropna(inplace = True)
song_stats.loc[:,features_list].describe()
```

```python
num_features = song_stats.loc[:,['BPM', 'ENERGY', 'DANCE', 'LOUD', 'VALENCE', 'ACOUSTIC', 'POP.']]
num_features.dtypes
```

```python
num_features
```

```python
vibe = song_stats.loc[:, ["Vibe"]]
```

```python
alfred_adalearn = AdaBoostClassifier()
```

```python
Xtrain, XTest, yTrain, yTest = train_test_split(num_features, vibe, test_size = 0.33)
```

```python
alfred_adalearn.fit(Xtrain, yTrain)
```

```python
ypred = alfred_adalearn.predict(XTest)
```

```python
print('Alfred acc:', accuracy_score(yTest, ypred))
```
